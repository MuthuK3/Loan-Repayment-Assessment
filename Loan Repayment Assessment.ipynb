```python
from sklearn.model_selection import train_test_split

X_num = df[numerical_columns].drop(columns=['loan_status_encoded'])
y = df['loan_status']

X_train, X_test, y_train, y_test = train_test_split(X_num, y, test_size=0.2, random_state=42)
```

```python
from sklearn.feature_selection import SelectKBest, f_classif

k = 5
selector_numerical = SelectKBest(score_func=f_classif, k=k)
X_train_numerical_selected = selector_numerical.fit_transform(X_train, y_train)

selected_indices_numerical = selector_numerical.get_support(indices=True)

selected_features_numerical = X_train.columns[selected_indices_numerical]

print("Selected Numerical Features:")
print(selected_features_numerical)
```

```python
from sklearn.model_selection import train_test_split

categorical_columns_without_target = [col for col in categorical_columns if col != 'loan_status']

X_cat = df[categorical_columns_without_target]
y = df['loan_status']

X_train, X_test, y_train, y_test = train_test_split(X_cat, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse=False)

X_encoded = encoder.fit_transform(X_cat)

encoded_columns = encoder.get_feature_names_out(categorical_columns_without_target)

X_encoded_df = pd.DataFrame(X_encoded, columns=encoded_columns)

X_final = pd.concat([X_cat, X_encoded_df], axis=1)

X_final.drop(columns=categorical_columns_without_target, inplace=True)

print("Final DataFrame with One-Hot Encoded Columns:")
print(X_final.head())
```

```python
from sklearn.feature_selection import SelectKBest, chi2

k = 5
selector_categorical = SelectKBest(score_func=chi2, k=k)
X_categorical_selected = selector_categorical.fit_transform(X_encoded_df, y)

selected_indices_categorical = selector_categorical.get_support(indices=True)

selected_features_categorical = X_encoded_df.columns[selected_indices_categorical]

print("Selected Categorical Features:")
print(selected_features_categorical)
```

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
X = pd.concat([X_num, X_final], axis=1)
y = df['loan_status_encoded']
```

```python
from imblearn.over_sampling import RandomOverSampler

oversampler = RandomOverSampler(random_state=42)

X_resampled, y_resampled = oversampler.fit_resample(X, y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

y_pred = rf_model.predict(X_test)

from sklearn.metrics import classification_report, accuracy_score
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

scores = cross_val_score(rf_model, X, y, cv=5)

print("Cross-validation scores:", scores)
print("Mean CV accuracy:", scores.mean())
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

classifiers = {
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Logistic Regression': LogisticRegression(max_iter=1000)
}

for name, clf in classifiers.items():
    scores = cross_val_score(clf, X, y, cv=5)
    print(f"{name} - Mean CV accuracy: {scores.mean()}")
```

```python
from sklearn.metrics import f1_score
from sklearn.model_selection import cross_val_predict

for name, clf in classifiers.items():
    y_pred = cross_val_predict(clf, X, y, cv=5)
    f1 = f1_score(y, y_pred)
    print(f"{name} - F1-score: {f1}")
```
